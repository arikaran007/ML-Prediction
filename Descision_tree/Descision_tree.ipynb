```python
import pandas as pd
df = pd.read_csv('/content/student-por.csv')
df.head()
```

```python
len(df)
```

```python
df['pass']  = df.apply(lambda row: 1 if (row['G1']+row['G2']+row['G3']) >=35 else 0,axis=1)
df=df.drop(['G1','G2','G3'],axis=1)
df.head()
```

```python
#ONE HOT ENCODING

df=pd.get_dummies(df,columns=['school','sex','address','famsize','Pstatus','Mjob','Fjob','internet','romantic','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher'])
df.head()
```

```python
x = df.drop(['pass'],axis=1)
x
```

```python
y = df['pass']
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn import tree
t = tree.DecisionTreeClassifier(criterion = 'entropy')
t =t.fit(x_train,y_train)
```

```python
t.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(t,x,y)
print(score.mean())
```

```python
# Make predictions on the test set
predictions = t.predict(x_test)

# Print the predictions
print(predictions)
```
